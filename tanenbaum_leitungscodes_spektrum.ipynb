# Visualisierung von Leitungscodes (Basisbandcodes)

In diesem Notebook werden verschiedene Leitungscodes im Basisband visualisiert, um die Umsetzung einer Bitfolge in ein physikalisches Leitungssignal besser zu verstehen.

Wir betrachten die folgenden Codierungen:

- **NRZ**
- **NRZI**
- **Manchester**
- **AMI**

Ziel ist es, zu sehen,

- wie aus einer Bitfolge ein zeitlicher Signalverlauf entsteht,
- welche Codierungen vom vorherigen Zustand abhängen,
- und warum sich die verschiedenen Verfahren in Bezug auf Übergänge, Synchronisation und Gleichanteil unterscheiden.

Oben wird jeweils die **Bitfolge** angezeigt, darunter der zugehörige **Signalverlauf**.
Zusätzlich kann eine **Erklärungstabelle** eingeblendet werden, die die Codierungsregel bitweise nachvollziehbar macht.

# Bedienung des interaktiven Notebooks

## Eingabefelder

- **Bits:**  
  Hier kann eine beliebige Bitfolge eingegeben werden, z. B.  
  `10110010`  
  (nur 0 und 1 erlaubt)

- **Code:**  
  Auswahl des Codierungsverfahrens:
  - NRZ
  - NRZI
  - Manchester
  - AMI

- **Startpegel (nur bei NRZI relevant):**  
  Legt fest, mit welchem Pegel das Signal beginnt (`+1` oder `-1`)

## Optionen

- **Erklärungstabelle anzeigen:** blendet die bitweise Herleitung unterhalb der Grafik ein.
- **Alle Codes vergleichen:** zeigt für dieselbe Bitfolge alle vier Codierungen untereinander.
- **Spektrum anzeigen:** ergänzt unter dem Zeitverlauf eine spektrale Darstellung des Signals.
- **DC-Anteil entfernen:** zieht für die Spektrumsdarstellung den Mittelwert ab, damit kleinere Spektralanteile besser sichtbar werden.

## Darstellung

- **Oben:** Bitfolge (ein Kästchen pro Bit)
- **Mitte:** resultierendes Signal über der Zeit
- **Unten (optional):** Spektrum des codierten Signals

- Vertikale Linien markieren **Bitgrenzen**
- Bei Manchester zusätzlich gestrichelte Linien für **Bitmitten**

## Tipps zur Nutzung

- Probiere verschiedene Bitfolgen aus, z. B.:
  - `111111` → wenige Übergänge bei NRZ
  - `101010` → viele Übergänge
  - `000000` → Verhalten bei konstantem Signal
- Schalte zwischen Zeit- und Spektraldarstellung um und vergleiche die Codes direkt.


## Wie ist die Darstellung zu lesen?

Die Visualisierung besteht aus zwei Teilen:

### Bitfolge
Im oberen Bereich ist die eingegebene Bitfolge dargestellt.  
Jedes Kästchen entspricht **einem Bitintervall**.

### Signalverlauf
Im unteren Bereich ist das codierte elektrische Signal als Funktion der Zeit dargestellt.  
Die horizontale Achse zeigt die Zeit in **Bitintervallen**, die vertikale Achse den **Signalpegel**.

- Ein konstanter Abschnitt bedeutet: Der Pegel bleibt während dieses Zeitintervalls gleich.
- Ein Sprung bedeutet: Es findet ein Übergang statt.
- Bei **Manchester** besteht jedes Bit aus **zwei Halbbits**, deshalb treten dort zusätzliche Übergänge innerhalb eines Bitintervalls auf.

### Erklärungstabelle
Die Tabelle unter dem Plot zeigt für jedes Bit, wie der zugehörige Pegel entsteht.
Bei zustandsabhängigen Verfahren wie **NRZI** und **AMI** ist diese Tabelle besonders hilfreich.

## NRZ (Non-Return-to-Zero)

Beim **NRZ-Code** wird jedem Bit direkt ein Pegel zugeordnet.  
In diesem Notebook verwenden wir die Zuordnung:

- **1 → +1**
- **0 → -1**

Das Signal bleibt während des gesamten Bitintervalls konstant.  
Es gibt also **keine Rückkehr auf Null innerhalb eines Bits**.

### Eigenschaften
- einfache Darstellung
- leicht zu erzeugen
- bei langen Folgen gleicher Bits entstehen nur wenige oder gar keine Übergänge

### Wichtig
NRZ zeigt sehr gut das Problem der **Taktrückgewinnung**:
Wenn über längere Zeit kein Übergang auftritt, wird es schwieriger, die Bitgrenzen zuverlässig zu erkennen.

## NRZI (Non-Return-to-Zero Inverted)

Beim **NRZI-Code** beschreibt das Bit **nicht direkt den Pegel**, sondern ob ein Übergang stattfindet.

In dieser Variante gilt:

- **1 → Pegelwechsel**
- **0 → kein Pegelwechsel**

Daher hängt das Signal nicht nur vom aktuellen Bit ab, sondern auch vom **vorherigen Pegel**.
Aus diesem Grund braucht NRZI immer einen **Startpegel**.

### So liest man die Tabelle
- **Vorheriger Pegel**: Pegel vor dem aktuellen Bit
- **Übergang**: Gibt an, ob ein Wechsel stattfindet
- **Neuer Pegel**: Pegel während des aktuellen Bitintervalls

### Eigenschaften
- zustandsabhängige Codierung
- mehr Übergänge als bei rein direkter Pegelzuordnung
- besonders geeignet, um Folgen von Einsen gut sichtbar zu machen

### Wichtig
Der zentrale Unterschied zu NRZ ist:

> Bei NRZ codiert das Bit den Pegel.  
> Bei NRZI codiert das Bit die Änderung des Pegels.

## Manchester-Code

Beim **Manchester-Code** wird jedes Bit in **zwei Halbbits** aufgeteilt.  
In der Mitte jedes Bits findet immer ein Übergang statt.

In dieser Darstellung verwenden wir die Konvention:

- **1 → low-to-high**
- **0 → high-to-low**

Das bedeutet:
Ein Bit wird nicht nur durch den Pegel beschrieben, sondern durch die **Richtung des Übergangs in der Bitmitte**.

### Eigenschaften
- in jedem Bit gibt es einen Übergang
- sehr gut für die Taktrückgewinnung
- benötigt mehr Bandbreite als NRZ oder NRZI

### Wichtig
Manchester ist besonders anschaulich, weil man direkt erkennt:

- die **Bitgrenzen**
- die **Bitmitte**
- und den systematischen Übergang innerhalb jedes Bits

Dadurch ist der Code gut selbsttaktend.

## AMI (Alternate Mark Inversion)

Beim **AMI-Code** werden die Nullen als **0-Pegel** dargestellt.  
Die Einsen werden abwechselnd als positiver und negativer Puls codiert:

- **0 → 0**
- **1 → abwechselnd +1, -1, +1, -1, ...**

Die Einsen wechseln also immer das Vorzeichen.
Dadurch hängt das Signal auch hier vom Verlauf der vorherigen Bits ab.

### Eigenschaften
- Nullen liegen auf dem Nullpegel
- Einsen alternieren zwischen positiv und negativ
- der Gleichanteil wird reduziert

### Wichtig
AMI zeigt schön, dass eine Codierung nicht nur zur Bitdarstellung dient, sondern auch Vorteile (etwa ein günstigeres Verhalten bezüglich Mittelwert) haben kann.

```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
import pandas as pd
from IPython.display import display, clear_output
```

```python
def parse_bits(bitstring):
    bitstring = bitstring.strip().replace(" ", "")
    if not bitstring:
        raise ValueError("Bitte eine Bitfolge eingeben.")
    if any(b not in "01" for b in bitstring):
        raise ValueError("Die Bitfolge darf nur aus 0 und 1 bestehen.")
    return [int(b) for b in bitstring]


def encode_nrz(bits, one_level=1, zero_level=-1):
    levels = [one_level if b == 1 else zero_level for b in bits]
    info = []
    for i, b in enumerate(bits):
        info.append({
            "Bitindex": i,
            "Bit": b,
            "Pegel": levels[i]
        })
    return levels, info


def encode_nrzi(bits, start_level=1, toggle_on="1"):
    level = start_level
    levels = []
    info = []

    for i, b in enumerate(bits):
        toggle = (b == 1) if toggle_on == "1" else (b == 0)

        old_level = level
        if toggle:
            level *= -1

        levels.append(level)
        info.append({
            "Bitindex": i,
            "Bit": b,
            "Vorheriger Pegel": old_level,
            "Übergang": "ja" if toggle else "nein",
            "Neuer Pegel": level
        })

    return levels, info


def encode_manchester(bits, convention="IEEE"):
    """
    IEEE-Konvention:
      0 -> high-to-low  => [+1, -1]
      1 -> low-to-high  => [-1, +1]
    """
    halfbit_levels = []
    info = []

    for i, b in enumerate(bits):
        if convention == "IEEE":
            pair = [-1, +1] if b == 1 else [+1, -1]
        else:
            pair = [+1, -1] if b == 1 else [-1, +1]

        halfbit_levels.extend(pair)
        info.append({
            "Bitindex": i,
            "Bit": b,
            "1. Halbbit": pair[0],
            "2. Halbbit": pair[1],
            "Mittelübergang": "immer"
        })

    return halfbit_levels, info


def encode_ami(bits):
    last_mark = -1
    levels = []
    info = []

    for i, b in enumerate(bits):
        if b == 0:
            level = 0
            mark = "-"
        else:
            last_mark *= -1
            level = last_mark
            mark = "+V" if level > 0 else "-V"

        levels.append(level)
        info.append({
            "Bitindex": i,
            "Bit": b,
            "Pegel": level,
            "Markierung": mark
        })

    return levels, info
```

```python
def step_plot(ax, levels, title="", ymin=None, ymax=None):
    n = len(levels)
    x = np.arange(n + 1)
    y = np.r_[levels, levels[-1]]
    ax.step(x, y, where="post")
    ax.set_xlim(0, n)
    ax.set_title(title)
    ax.set_xlabel("Bitzeit")
    ax.set_ylabel("Pegel")
    ax.grid(True, alpha=0.3)
    ax.axhline(0, linewidth=1)

    for k in range(n + 1):
        ax.axvline(k, linestyle=":", alpha=0.3)

    if ymin is not None and ymax is not None:
        ax.set_ylim(ymin, ymax)


def plot_manchester(ax, halfbit_levels, title="Manchester", ymin=-1.5, ymax=1.5):
    n_half = len(halfbit_levels)
    x = np.arange(n_half + 1) / 2.0
    y = np.r_[halfbit_levels, halfbit_levels[-1]]

    ax.step(x, y, where="post")
    ax.set_xlim(0, n_half / 2)
    ax.set_ylim(ymin, ymax)
    ax.set_title(title)
    ax.set_xlabel("Bitzeit")
    ax.set_ylabel("Pegel")
    ax.grid(True, alpha=0.3)
    ax.axhline(0, linewidth=1)

    n_bits = n_half // 2
    for k in range(n_bits + 1):
        ax.axvline(k, linestyle=":", alpha=0.4)
    for k in range(n_bits):
        ax.axvline(k + 0.5, linestyle="--", alpha=0.25)


def compute_spectrum(levels, samples_per_symbol=100, remove_dc=False):
    signal = np.repeat(np.asarray(levels, dtype=float), samples_per_symbol)

    if remove_dc:
        signal = signal - np.mean(signal)

    n = len(signal)
    freqs = np.fft.rfftfreq(n, d=1.0 / samples_per_symbol)
    magnitude = np.abs(np.fft.rfft(signal)) / n
    return freqs, magnitude


def plot_spectrum(ax, freqs, magnitude, title="Spektrum", max_freq_factor=0.2):
    ax.plot(freqs, magnitude)
    ax.set_title(title)
    ax.set_xlabel("Frequenz (normiert)")
    ax.set_ylabel("Amplitude")
    ax.grid(True, alpha=0.3)

    if len(freqs) > 1:
        ax.set_xlim(0, freqs[-1] * max_freq_factor)

    # DC-Spitze nicht die komplette Darstellung dominieren lassen
    if len(magnitude) > 1:
        visible = magnitude[1:] if len(magnitude) > 2 else magnitude
        ymax = float(np.max(visible)) if len(visible) else 0.0
        if ymax > 0:
            ax.set_ylim(0, ymax * 1.1)
```

```python
def draw_bit_boxes(ax, bits):
    n = len(bits)
    ax.set_xlim(0, n)
    ax.set_ylim(0, 1)
    ax.axis("off")

    for i, b in enumerate(bits):
        rect = plt.Rectangle((i, 0), 1, 1, fill=False, linewidth=1.2)
        ax.add_patch(rect)
        ax.text(i + 0.5, 0.5, str(b), ha="center", va="center", fontsize=12)

    for i in range(n):
        ax.text(i + 0.5, 1.05, f"{i}", ha="center", va="bottom", fontsize=9)


def print_info_table(info, max_rows=30):
    if not info:
        return

    df = pd.DataFrame(info)

    if len(df) > max_rows:
        display(df.head(max_rows))
        print(f"... ({len(df) - max_rows} weitere Zeilen)")
    else:
        display(df)
```

```python
def visualize_line_code(
    bitstring,
    code,
    start_level=1,
    show_info=True,
    compare_all=False,
    show_spectrum=True,
    remove_dc=True,
):
    bits = parse_bits(bitstring)

    if compare_all:
        if show_spectrum:
            fig, axes = plt.subplots(
                6, 1, figsize=(12, 12),
                gridspec_kw={"height_ratios": [0.8, 2, 2, 2, 2, 2.2]},
                constrained_layout=True,
            )
            ax_bits, ax_nrz, ax_nrzi, ax_man, ax_ami, ax_spec = axes
        else:
            fig, axes = plt.subplots(
                5, 1, figsize=(12, 10),
                gridspec_kw={"height_ratios": [0.8, 2, 2, 2, 2]},
                constrained_layout=True,
            )
            ax_bits, ax_nrz, ax_nrzi, ax_man, ax_ami = axes
            ax_spec = None

        draw_bit_boxes(ax_bits, bits)

        nrz_levels, _ = encode_nrz(bits)
        step_plot(ax_nrz, nrz_levels, title="NRZ", ymin=-1.5, ymax=1.5)

        nrzi_levels, _ = encode_nrzi(bits, start_level=start_level)
        step_plot(ax_nrzi, nrzi_levels, title=f"NRZI (Startpegel {start_level})", ymin=-1.5, ymax=1.5)

        man_levels, _ = encode_manchester(bits)
        plot_manchester(ax_man, man_levels, title="Manchester", ymin=-1.5, ymax=1.5)

        ami_levels, _ = encode_ami(bits)
        step_plot(ax_ami, ami_levels, title="AMI", ymin=-1.5, ymax=1.5)

        if ax_spec is not None:
            spectra = {
                "NRZ": compute_spectrum(nrz_levels, samples_per_symbol=100, remove_dc=remove_dc),
                "NRZI": compute_spectrum(nrzi_levels, samples_per_symbol=100, remove_dc=remove_dc),
                "Manchester": compute_spectrum(man_levels, samples_per_symbol=50, remove_dc=remove_dc),
                "AMI": compute_spectrum(ami_levels, samples_per_symbol=100, remove_dc=remove_dc),
            }
            for label, (freqs, mag) in spectra.items():
                ax_spec.plot(freqs, mag, label=label)
            ax_spec.set_title("Spektren im Vergleich")
            ax_spec.set_xlabel("Frequenz (normiert)")
            ax_spec.set_ylabel("Amplitude")
            ax_spec.grid(True, alpha=0.3)
            ax_spec.legend()
            all_freqs = [freqs for freqs, _ in spectra.values() if len(freqs) > 1]
            if all_freqs:
                max_freq = min(freqs[-1] for freqs in all_freqs)
                ax_spec.set_xlim(0, max_freq * 0.2)
            all_mags = [mag[1:] if len(mag) > 2 else mag for _, mag in spectra.values() if len(mag) > 1]
            if all_mags:
                ymax = max(float(np.max(m)) for m in all_mags if len(m))
                if ymax > 0:
                    ax_spec.set_ylim(0, ymax * 1.1)

        plt.show()
        return

    if show_spectrum:
        fig, axes = plt.subplots(
            3, 1, figsize=(12, 7),
            gridspec_kw={"height_ratios": [0.8, 3, 2]},
            constrained_layout=True,
        )
        ax_bits, ax_signal, ax_spec = axes
    else:
        fig, axes = plt.subplots(
            2, 1, figsize=(12, 5),
            gridspec_kw={"height_ratios": [0.8, 3]},
            constrained_layout=True,
        )
        ax_bits, ax_signal = axes
        ax_spec = None

    draw_bit_boxes(ax_bits, bits)

    info = None
    spectrum_levels = None
    samples_per_symbol = 100

    if code == "NRZ":
        levels, info = encode_nrz(bits)
        step_plot(ax_signal, levels, title="NRZ", ymin=-1.5, ymax=1.5)
        spectrum_levels = levels

    elif code == "NRZI":
        levels, info = encode_nrzi(bits, start_level=start_level)
        step_plot(ax_signal, levels, title=f"NRZI (Startpegel {start_level})", ymin=-1.5, ymax=1.5)
        spectrum_levels = levels

    elif code == "Manchester":
        levels, info = encode_manchester(bits)
        plot_manchester(ax_signal, levels, title="Manchester", ymin=-1.5, ymax=1.5)
        spectrum_levels = levels
        samples_per_symbol = 50

    elif code == "AMI":
        levels, info = encode_ami(bits)
        step_plot(ax_signal, levels, title="AMI", ymin=-1.5, ymax=1.5)
        spectrum_levels = levels

    else:
        raise ValueError("Unbekannter Code.")

    if show_spectrum and ax_spec is not None and spectrum_levels is not None:
        freqs, mag = compute_spectrum(
            spectrum_levels,
            samples_per_symbol=samples_per_symbol,
            remove_dc=remove_dc,
        )
        plot_spectrum(ax_spec, freqs, mag, title="Spektrum")

    plt.show()

    if show_info:
        print_info_table(info)
```

```python
bit_input = widgets.Text(
    value="10110010",
    description="Bits:",
    layout=widgets.Layout(width="500px")
)

code_dropdown = widgets.Dropdown(
    options=["NRZ", "NRZI", "Manchester", "AMI"],
    value="NRZ",
    description="Code:"
)

start_level_dropdown = widgets.Dropdown(
    options=[(" +1", 1), (" -1", -1)],
    value=1,
    description="Startpegel:"
)

show_info_checkbox = widgets.Checkbox(
    value=True,
    description="Erklärungstabelle anzeigen"
)

compare_all_checkbox = widgets.Checkbox(
    value=False,
    description="Alle Codes vergleichen"
)

show_spectrum_checkbox = widgets.Checkbox(
    value=True,
    description="Spektrum anzeigen"
)

remove_dc_checkbox = widgets.Checkbox(
    value=True,
    description="DC-Anteil im Spektrum entfernen"
)

output = widgets.Output()


def update_plot(*args):
    with output:
        clear_output(wait=True)
        try:
            visualize_line_code(
                bitstring=bit_input.value,
                code=code_dropdown.value,
                start_level=start_level_dropdown.value,
                show_info=show_info_checkbox.value,
                compare_all=compare_all_checkbox.value,
                show_spectrum=show_spectrum_checkbox.value,
                remove_dc=remove_dc_checkbox.value,
            )
        except Exception as e:
            print(f"Fehler: {e}")


bit_input.observe(update_plot, names="value")
code_dropdown.observe(update_plot, names="value")
start_level_dropdown.observe(update_plot, names="value")
show_info_checkbox.observe(update_plot, names="value")
compare_all_checkbox.observe(update_plot, names="value")
show_spectrum_checkbox.observe(update_plot, names="value")
remove_dc_checkbox.observe(update_plot, names="value")

controls = widgets.VBox([
    bit_input,
    widgets.HBox([code_dropdown, start_level_dropdown]),
    widgets.HBox([show_info_checkbox, compare_all_checkbox]),
    widgets.HBox([show_spectrum_checkbox, remove_dc_checkbox]),
])

display(controls, output)
update_plot()
```

## Vergleich der Codierungen

Wenn dieselbe Bitfolge mit verschiedenen Verfahren codiert wird, entstehen deutlich unterschiedliche Signalverläufe.

### NRZ
- direkte Pegelzuordnung
- einfach
- wenig Übergänge bei langen gleichen Bitfolgen

### NRZI
- Übergänge statt absoluter Pegel sind entscheidend
- abhängig vom Startzustand
- gut geeignet, um Änderungen sichtbar zu machen

### Manchester
- immer ein Übergang in der Bitmitte
- sehr gut für Synchronisation
- höherer Bandbreitenbedarf

### AMI
- Nullen bleiben auf Null
- Einsen wechseln das Vorzeichen
- reduziert den Gleichanteil

### Zentrale Erkenntnis
Dieselbe Bitfolge kann auf sehr unterschiedliche Weise auf ein Leitungssignal abgebildet werden.  
Die Wahl des Codes beeinflusst unter anderem:

- die Anzahl der Übergänge,
- die Synchronisierbarkeit,
- den Gleichanteil,
- und den Bandbreitenbedarf.
